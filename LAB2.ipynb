```python
d1 = pd.DataFrame ({'RN':[1,2,3,4,5],'SName':["Sandesh" , "Raj"," Ram" ,"Nishant","Swapnil"]})
d2 = pd.DataFrame ({'RN':[1,2,3,4,5],'Marks':[20,25,23,19,29]})
```
